```python
#Converting Sentences into embeddings
```

```python
f = open('glove.6B.50d.txt',encoding='utf-8')
```

```python
embeddings_index = {}
count = 0
for line in f:
    values = line.split()
    word = values[0]
    coefs = np.asarray(values[1:],dtype='float')
    
    embeddings_index[word] = coefs#Here it is building up a dictionary for each word
f.close()
    
```

```python
embeddings_index["eat"]
```

```python
emb_dim = embeddings_index["eat"].shape[0]
print(emb_dim)
```

```python
###Convert our sentence into the vectors
```

```python
def embedding_output(X):
    maxLen = 10
    embedding_out = np.zeros((X.shape[0],maxLen,emb_dim))
    
    for ix in range(X.shape[0]):
        X[ix] = X[ix].split()
        for ij in range(len(X[ix])):
            #go to every word in current ix sentence
            try:
                embedding_out[ix][ij] = embeddings_index[X[ix][ij].lower()]
            except:
                embedding_out[ix][ij] = np.zeros((50,))
    
    return embedding_out
            
            
```

```python
embedding_matrix_train = embedding_output(X_train)
embedding_matrix_test = embedding_output(X_test)
```

```python
print(X_train[0])
print(len(X_train[0]))
```

```python
print(embedding_matrix_train.shape)
print(embedding_matrix_test.shape)
```

```python
from keras.utils import to_categorical
```

```python
Y_train = to_categorical(Y_train,num_classes=5)
Y_test = to_categorical(Y_test,num_classes=5)


#We need to convert Y_train,test in one hot encoder
```

```python
print(Y_train[0])
```

```python
#Define the rnn/lstm model

from keras.models import Sequential
from keras.layers import *
```

```python
model = Sequential()
model.add(LSTM(64,input_shape=(10,50),return_sequences=True))
model.add(Dropout(0.5))
model.add(LSTM(64,return_sequences=False))
model.add(Dropout(0.5))
model.add(Dense(5))
model.add(Activation('softmax'))
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
model.summary()
```

```python
from keras.callbacks import EarlyStopping
from keras.callbacks import ModelCheckpoint

checkpoint = ModelCheckpoint("./emojimodel/emoji_best_model.h5",monitor='val_loss',verbose=True,save_best_only=True)
earlystop = EarlyStopping(monitor='val_accuracy',patience=10)
hist = model.fit(embedding_matrix_train,Y_train,epochs=60,batch_size=64,shuffle=True,validation_split=0.2)
```

```python
pred = model.predict_classes(embedding_matrix_test)
```

```python
model.evaluate(embedding_matrix_test,Y_test)
```
